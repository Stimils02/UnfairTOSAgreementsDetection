## All at once

```python
import pandas as pd
import os

base_path = '/Users/noshitha/Downloads/terms'

# Loop through all months and filter
for month in range(1, 13):
    filename = f"2024_{str(month).zfill(2)}_terms.csv"
    file_path = os.path.join(base_path, filename)
    
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        filtered = df[
            (df['website_lang'] == 'uk') &
            (df['terms_content'].notna()) &
            (df['terms_lang'] == 'en')
        ][['website_month_id', 'website_lang', 'website_url', 'terms_url',
           'terms_content', 'terms_lang', 'terms_keyword_score', 'terms_ml_probability']]
        
        output_filename = f"filtered_uk_terms_{str(month).zfill(2)}.csv"
        output_path = os.path.join(base_path, output_filename)
        filtered.to_csv(output_path, index=False)
        print(f"Saved filtered file: {output_filename} with {len(filtered)} rows")
    else:
        print(f"File not found: {file_path}")
```

```python
# Collect all filtered files
filtered_files = [f"filtered_uk_terms_{str(month).zfill(2)}.csv" for month in range(1, 13)]

# Read and concatenate
all_filtered = []
for filename in filtered_files:
    file_path = os.path.join(base_path, filename)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        all_filtered.append(df)
    else:
        print(f"Missing filtered file: {filename}")

# Combine and save
combined_df = pd.concat(all_filtered, ignore_index=True)
combined_path = os.path.join(base_path, 'filtered_uk_terms_all_months.csv')
combined_df.to_csv(combined_path, index=False)

print(f"Final combined file saved with {len(combined_df)} rows at: {combined_path}")
```
